```python
import torch
from torchvision import models
```

```python
if torch.cuda.is_available():
    device = "cuda"
else:
    device = "cpu"
```

```python
resnet_50 = models.resnet50(weights=models.ResNet50_Weights.DEFAULT)
print(resnet_50)
```

```python
torch.save(resnet_50.state_dict(), 'resnet_50.pt')
```

```python
resnet_50.load_state_dict(torch.load('resnet_50.pt'))
```
